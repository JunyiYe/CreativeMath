```python
def extract_yes_no(cell):
    # Check if the cell is NaN and return "N/A"
    if pd.isna(cell):
        return "N/A"
    # Check if "YES" or "NO" is in the cell and return it, otherwise return an empty string
    elif "YES" in cell:
        return "YES"
    elif "NO" in cell:
        return "NO"
    else:
        return ""
```

```python
import numpy as np

num_samples = 17
# Initialize a similarity matrix
similarity_matrix = pd.DataFrame(1, index=models, columns=models)

# Calculate similarity scores
for i in range(num_models):
    for j in range(i + 1, num_models):
        model1, model2 = models[i], models[j]
        similar = df[(df['Model_i'] == model1) & (df['Model_j'] == model2)]['Novelty_major_voting'].value_counts().get('NO', 0)
        similarity_score = similar / num_samples
        similarity_matrix.at[model1, model2] = similarity_score
        similarity_matrix.at[model2, model1] = similarity_score


# Correct the diagonal filling
np.fill_diagonal(similarity_matrix.values, 0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(similarity_matrix, annot=True, cmap='Greens', fmt=".2f")
plt.title('Model Pairwise Similarity Scores')
plt.show()
```

```python
# Calculate the mean similarity for each model, excluding self-comparison (diagonal)
mean_similarity = similarity_matrix.mean(axis=1)  # You can use axis=0 or axis=1 since the matrix is symmetric

# Display the results
print("Mean Similarity Scores for Each Model:")
print(mean_similarity)
```
